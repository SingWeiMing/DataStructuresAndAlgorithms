### a. Briefly describe two main differences between Big O notation and Big Omega  notation. (6 marks)

One difference between the two is that Big O notation is meant to represent a worst case scenario,  while Big Omega notation is meant to represent the best case scenario, the fastest possible outcome.
For example, in quicksort in terms of time complexity, its Big O is O(n^2), while its Big Omega is (n log(n)).

Another difference is that in Big O notation the rate of growth of an algorithm is less than or equal to the behaviour in O(), while in Big Omega notation the rate of growth of an algorithm is more than or equal to the behaviour in ().

### b. Under what situation can we ignore big O notation? ( 1.5 marks)

### b. Quick sort has a worst case time complexity of big O(n<sup>2</sup>). Explain the worst case scenario and make a suggestion to avoid running into the situation. (2 marks)

Answer: The worst case scenario is when the list is already sorted, and the quicksort takes the right most/ left most as a pivot. The sorting would just keep going one element at a time till it reaches the end of the list before doing the same to gather back the list as a sorted list.

A way to avoid running into this situation would be to have the pivot be from the middle of the lists, as this would allow the lists to be split with a almost identical amount of elements.

### A program is required to record and display the results of a swimming competition. The swimming pool has 5 lanes. The swimmer name for each lane is first recorded. After the race, the timings for the swimmers are captured. The results for the race is then displayed.


### a.	Write a function inputSwimmers() that prompts for 5 swimmers' names and returns the names in a list. An example of the input session is as follows: ( 3 marks)
#### Enter lane 1 swimmer: Jin Jia Ho
#### Enter lane 2 swimmer: Mai Siao Siao
#### 
#### Enter lane 5 swimmer: Ho Seh Lah



```python
def inputSwimmers():
    '''The function can be called to receive 5 name inputs from the user, before returning the inputs as a list.'''
    name_list_temp=[] #creating a new temp list
    
    for i in range(0, 5): #for loop that starts at 0 and stops at 4, just before 5.
        i = str(i+1) #making i one number larger and converting it into a string so it can be concatenated.
        ele = input("Enter lane "+i+" swimmer:") #user inputs the swimmer names
        name_list_temp.append(ele)#adds the input to the temp list, then the loop starts again
        
    return name_list_temp    
```

### b.	Write another function inputTiming(swimmers) that has the list of swimmers names as paramter. The function creates an empty list, prompts and stores the timing for each of the swimmers. The timing list is returned. An example of the input session is as follows: (3 marks)
#### Enter timing for Jin Jia Ho: 47.15
#### Enter timing for Mai Siao Siao: 46.91
#### .
#### Enter timing for Ho Seh Lah: 48.01


```python
def inputTiming(name):
    '''The function can be called to receive as many timings as 
    there are names from the user, before returning the inputs as a list'''
    timing_list_temp=[]#creating a new temp list
    
    for i in name: #for loop that iterates for as many elements as there are in name
        ele = float(input("Enter timing for "+i+":")) 
        #user inputs the timings for the swimmers, and the input is converted to a float.
        timing_list_temp.append(ele) #the input is added to the temp list, then the loop starts again.
        
    return timing_list_temp
```

#### c.	Write another function printResults(swimmers, timings) that has the name and timing lists as parameters. The function prints a summary of the results as follows: (3 marks)
#### Jin Jia Ho           47.15s
#### Mai Siao Siao     46.91s
#### 
#### Ho Seh Lah        48.01s
#### The Fastest timing is 46.91s


```python
def printResults(swimmers,timing):
    '''Prints the results in order of the name input.'''
    for i in range(len(swimmers)): #for loop that iterates for as many elements that is in the swimmers list.
        print(swimmers[i],' ',timing[i],'s',sep='')
        #prints the swimmers and timings of the index number of the loop, and concatenates a 's' at timing for seconds.
    print('The fastest timing is ',min(timing),'s',sep='')
```

### d.	Modify the function in part (c) and name it printResults2(swimmers, timings) that prints the summary in sorted ascending order of the timing. Therefore, the summary may look like this: (5 marks)
#### Mai Siao Siao	46.91s
#### Jin Jia Ho	 47.15s
#### 
#### Ho Seh Lah	48.01s

```python
def printResults2(swimmers,timing): 
    '''Prints the results in ascending order.'''
    timing, swimmers = zip(*sorted(zip(timing, swimmers))) 
    # the timings and swimmers are zipped to make a tuple to correspond the timings and swimmers, 
    #and are unpacked and sorted by the timing, then zipped again.
    for i in range(len(swimmers)): #for loop that iterates for as many elements that is in the swimmers list.
        print(swimmers[i],' ',timing[i],"s",sep='') #prints the swimmers and timings of the index number of the loop
    
```
